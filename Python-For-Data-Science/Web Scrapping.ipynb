```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
url = 'https://www.imdb.com/chart/boxoffice?ref_=nv_ch_cht_1'
```

```python
html = requests.get(url)
```

```python
bsObj = BeautifulSoup(html.text, 'lxml')
```

```python
bsObj
```

```python
bsObj.find('title').getText()
```

```python
h1_list = bsObj.findAll('h1')
```

```python
h1_list
```

```python
h4_list = bsObj.findAll('h4')
```

```python
for h4 in h4_list:
    print(h4.getText())
```

```python
title_list = bsObj.findAll('td', {'class': 'titleColumn'})
```

```python
for title in title_list:
    print(title.a.getText())
```
